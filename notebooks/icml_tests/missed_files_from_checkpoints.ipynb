```python
import scipy as sp
import numpy as np
import pandas as pd
import pickle
import torch
import os

from tqdm.notebook import tqdm
```

```python
folder_path = "../../runs/scripts/checkpoints/"
checkpoints = [folder_path + entry.name for entry in os.scandir(folder_path) if entry.is_file()]
```

```python
folder_path = "./sweep_results/"
files = [folder_path + entry.name for entry in os.scandir(folder_path) if (entry.is_file() and "loc4" in entry.name)]
```

```python
for checkpoint in tqdm(checkpoints):
    params = checkpoint.split("/")[-1][:-4].split("_")[1:]
    file = folder_path + f"sweep_loc4_{params[0]}_{params[1]}_{params[2]}.pkl"

    if file in files:
        continue
    else:
        print(file)
        print(checkpoint)
        with open(checkpoint, "rb") as f:
            print(f)
            saved_vars = pickle.load(f)

        results = saved_vars["results"]

        with open(file, "wb") as f:
            pickle.dump(results, f)
```
