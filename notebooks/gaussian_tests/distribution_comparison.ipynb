```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(1, 3, figsize=(5*3, 5))

for idx, k in enumerate([1, 2, 4]):
    x1 = np.random.randn(2000, 2)
    x2 = k + np.random.randn(2000, 2)
    x = np.concatenate([x1, x2], axis=0)

    sns.kdeplot(x=x[:, 0], y=x[:, 1], cmap="viridis", fill=True, ax=axes[idx])
    axes[idx].set_title(f"Location: {k}", fontsize=16)
    axes[idx].tick_params(labelsize=16)

plt.tight_layout()
plt.savefig("saved_plots/overlap_location.svg", dpi=300, bbox_inches="tight")
plt.show()
```
