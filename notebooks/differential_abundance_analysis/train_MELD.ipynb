```python
import meld 
import numpy as np
import os
import pandas as pd
from pathlib import Path
import scanpy as sc
```

Here we run the whole MELD pipeline. 

```python
data_path = Path("../../project_folder/data/pbmc68k_oversampled/")
res_dir = Path("../../project_folder/results/abundance_test_experiment/meld/")
```

```python
for adata_name in os.listdir(data_path):
    # Read adata 
    adata = sc.read_h5ad(data_path / adata_name)
    # Run MELD 
    X = adata.obsm["X_pca"]
    labels = adata.obs["treatment"].values
    sample_densities = meld.MELD().fit_transform(X, labels)
    log_ratio = np.log(sample_densities[1]) - np.log(sample_densities[0])
    adata.obs["log_ratio"] = log_ratio.values
    adata.write_h5ad(res_dir / f"oversamp_{adata_name.split('_')[1]}")
```
