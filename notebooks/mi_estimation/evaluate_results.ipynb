```python
import numpy as np
from pathlib import Path 
import matplotlib.pyplot as plt
import os
import pandas as pd
```

## Get metrics for the normal scRatio

```python
dims = np.array([20, 40, 80, 160, 320])
gt = 0.5 * np.log(1/(0.36**(dims/2)))
true_mi = dict(zip(dims, gt))
```

```python
result_dir = Path("../../project_folder/results/mi_experiments/scRatio/simulate_under_uncond/")
results = {"Config": [], 
           "MSE": [],
           "Dimensions": [], 
           "Run": []
          }

for file in os.listdir(result_dir):
    for dim in [20, 40, 80, 160, 320]:
        ratios = np.load(result_dir / file / f"ratios_{dim}.npy")
        ratios_split = np.split(ratios, 3)
        split_avgs = [arr.mean() for arr in ratios_split]
        # MSE with real values 
        split_avgs_mse = [np.abs(avg - true_mi[dim]) for avg in split_avgs]
        for i, avg in enumerate(split_avgs_mse):
            results["Config"].append(file)
            results["MSE"].append(avg)
            results["Dimensions"].append(dim)
            results["Run"].append(i)
```

```python
results_df = pd.DataFrame(results)
results_df.groupby(["Config", "Dimensions"]).mean()
```

```python
results_df.groupby(["Config", "Dimensions"]).std() / np.sqrt(3)
```

## Results baseline DRECFM

Results for the competing models were obtained using the code at https://github.com/ksnxr/dre-prob-paths using the "TwoSB" option. The adapted repo will be made available upon request, message alessandro.palma@helmholtz-munich.de.

```python
path_dre_cfm = Path("../../project_folder/results/mi_experiments/dre-cfm/dre-cfm_sb/")
```

```python
results_baseline = {"MSE": [],
                    "Dimensions": [], 
                    "Run": []
                      }
```

```python
for file in os.listdir(path_dre_cfm):
    _, dimensions, run = file.split("_")  
    dimensions = int(dimensions)
    run = int(run)
    results_baseline["Run"].append(run)
    results_baseline["Dimensions"].append(dimensions)
    
    ratios_cfm = np.load(path_dre_cfm / file / "metrics" / "estim_logr.npy")
    ratio_mean = ratios_cfm.mean()
    results_baseline["MSE"].append(np.abs(ratio_mean - true_mi[dimensions]))
```

```python
results_baseline_df = pd.DataFrame(results_baseline)
```

```python
results_baseline_df.groupby("Dimensions").mean()
```

```python
results_baseline_df.groupby("Dimensions").std() / np.sqrt(3)
```
