```python
from pathlib import Path 
import sys
import os 
import numpy as np
import scipy as sp
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import seaborn as sns
import matplotlib.pyplot as plt
```

## Path to result 

```python
dims = np.array([20, 40, 80, 160, 320])

gt = 0.5 * np.log(1/(0.36**(dims/2)))
true_mi = dict(zip(dims, gt))
```

```python
results_tsm = {"dimensions": [], 
              "lr": [], 
              "sb_var": [], 
              "mse": []}

results_ctsm = {"dimensions": [], 
              "lr": [], 
              "sb_var": [], 
              "mse": []}
```

### Results TSM

```python
result_dir_tsm = Path("../../project_folder/results/mi_experiments/dre-cfm/dre-cfm_tsm_sweeps/")
```

```python
for run in os.listdir(result_dir_tsm):
    run_split = run.split("_")
    dimension, lr, sb_var = int(run_split[4]), float(run_split[6]), float(run_split[-1])
    if sb_var==0:
        continue
    results_tsm["dimensions"].append(dimension)
    results_tsm["lr"].append(lr)
    results_tsm["sb_var"].append(sb_var)

    ratio = np.load(result_dir_tsm / run / "metrics" / "estim_logr.npy")
    
    ratio_mean = ratio.mean()
    results_tsm["mse"].append(np.abs(ratio_mean - true_mi[dimension]))
```

```python
results_tsm_df = pd.DataFrame(results_tsm)
```

```python
results_tsm_df.groupby(["dimensions", "lr", "sb_var"]).mean()
```

Best configurations to retrain TSM: 

Dimension 20: 
* lr 1e-3
* sb_var 0.5

Dimension 40:
* lr 5e-4
* sb_var 0.5

Dimension 80:
* lr 5e-3
* sb_var 0.1

Dimension 160:
* lr 5e-4
* sb_var 0.1

Dimension 320: 
* lr 5e-4
* sb_var 0.1

### Results CTSM

```python
result_dir_ctsm = Path("../../project_folder/results/mi_experiments/dre-cfm/dre-cfm_sb_sweeps")
```

```python
for run in os.listdir(result_dir_ctsm):
    run_split = run.split("_")
    dimension, lr, sb_var = int(run_split[4]), float(run_split[6]), float(run_split[-1])
    results_ctsm["dimensions"].append(dimension)
    results_ctsm["lr"].append(lr)
    results_ctsm["sb_var"].append(sb_var)

    ratio = np.load(result_dir_ctsm / run / "metrics" / "estim_logr.npy")
    
    ratio_mean = ratio.mean()
    results_ctsm["mse"].append(np.abs(ratio_mean - true_mi[dimension]))
```

```python
results_ctsm_df = pd.DataFrame(results_ctsm)
```

```python
results_ctsm_df.groupby(["dimensions", "lr", "sb_var"]).mean()
```

## Plot performance vs stochasticity

```python
sns.set_theme(style="whitegrid", context="talk")

plt.figure(figsize=(6,4))

ax = sns.boxplot(
    data=results_ctsm_df,
    x="dimensions",
    y="mse",
    hue="sb_var",
    width=0.6,
    fliersize=3,
    linewidth=1.2
)

ax.set_xlabel("Dimensions")
ax.set_ylabel("MSE")

# Legend in top-left
ax.legend(
    title="SB Variance",
    loc="upper left",
    fontsize=12,        # legend text size
    title_fontsize=13  # legend title size
)

sns.despine()
plt.tight_layout()
plt.title("CTSM", fontsize=15)
plt.savefig("../../project_folder/plots/mse_boxplot_ctsm.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
