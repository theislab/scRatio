# Split data 

```python
from sklearn.model_selection import train_test_split

n_samples_per_dimension = {}
for dim in [20, 40, 80, 160, 320]:
    correlated_gaussians_model = CorrelatedGaussianMI(dim)
    # Correlated Gaussian
    p_samples = correlated_gaussians_model.sample_p(100_000).numpy()
    # Standard Gaussian 
    q_samples = correlated_gaussians_model.sample_q(100_000).numpy()
    
    for samples, name in [(p_samples, "block_sigma"), (q_samples, "identity_sigma")]:
        # Split off test set (10%)
        train_val, test = train_test_split(samples, test_size=0.1, random_state=42)
        # Split remaining into train (80%) and validation (10%)
        train, val = train_test_split(train_val, test_size=0.1/0.9, random_state=42)
        
        np.save(dest_path / "split_data" / f"{name}_{dim}_train.npy", train)
        np.save(dest_path / "split_data" / f"{name}_{dim}_val.npy", val)
        np.save(dest_path / "split_data" / f"{name}_{dim}_test.npy", test)
```
