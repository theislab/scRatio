# Now we generate and save the data for different dimensionalities

```python
dest_path = Path("/home/icb/alessandro.palma/environment/scFM_density_estimation/project_folder/data/mi_data/")
```

```python
n_samples_per_dimension = {}

for dim in [20, 40, 80, 160, 320]:
    correlated_gaussians_model = CorrelatedGaussianMI(dim)
    # Correlated Gaussian
    p_samples = correlated_gaussians_model.sample_p(100_000).numpy()
    # Standard Gaussian 
    q_samples = correlated_gaussians_model.sample_q(100_000).numpy()
    
    np.save(dest_path / f"block_sigma_{dim}.npy", p_samples)
    np.save(dest_path / f"identity_sigma_{dim}.npy", q_samples)
```
