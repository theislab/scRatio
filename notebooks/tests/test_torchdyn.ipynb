```python
import torch
import numpy as np
from torchdyn.core import NeuralODE
```

```python
class NODEWrapper(torch.nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, t, x, *args, **kwargs):
        return t
```

```python
node = NeuralODE(NODEWrapper(), solver="dopri5", sensitivity="adjoint", atol=1e-4, rtol=1e-4)
```

```python
with torch.no_grad():
    traj = node.trajectory(
        torch.zeros((1, 1)),
        t_span=torch.linspace(1, 0, 100)
    )
```
